```python
covid_df.drop(["Sno","ConfirmedIndianNational","ConfirmedForeignNational"],inplace=True, axis=1)
```

```python
covid_df.tail() #i reruned it.
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = "%Y-%m-%d")
```

```python
covid_df.tail()
```

```python
#Active Cases
covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
```

```python
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values=['Confirmed','Deaths','Cured'], index = "State/UnionTerritory", aggfunc=max)
statewise
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
statewise
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
statewise
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending=False)
statewise
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
# top 10 active cases

top_10_active_cases = covid_df.groupby(by="State/UnionTerritory").max()[["Active_Cases","Date"]].sort_values(by=["Active_Cases"], ascending=False).reset_index()
top_10_active_cases
```

```python
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with highest Active Cases", size=25)
```

```python
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_Cases", x = "State/UnionTerritory",  linewidth = 2, edgecolor = "red")
```

```python
top_10_active_cases = covid_df.groupby(by="State/UnionTerritory").max()[["Active_Cases","Date"]].sort_values(by=["Active_Cases"], ascending=False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with highest Active Cases", size=25)
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_Cases", x = "State/UnionTerritory",  linewidth = 2, edgecolor = "red")
plt.xlabel("States")
plt.ylabel("Total Active Cases")
# ax.set_xticklabels(rotation=45) error
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

```python
#top 10 states with highest deaths

top_10_deaths = covid_df.groupby(by = "State/UnionTerritory").max()[["Deaths","Date"]].sort_values(by=["Deaths"],ascending = False).reset_index()
top_10_deaths
```

```python
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with Highest Deaths", size=25)
```

```python
top_10_deaths = covid_df.groupby(by = "State/UnionTerritory").max()[["Deaths","Date"]].sort_values(by=["Deaths"],ascending = False).reset_index()
fig = plt.figure(figsize=(18,5))
plt.title("Top 10 States with Highest Deaths", size=25)
ax = sns.barplot(data = top_10_deaths.iloc[:12], y="Deaths", x="State/UnionTerritory", linewidth = 2,edgecolor = "black")
plt.xlabel("State")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
#Remove all the data where State=India

vaccine = vaccine_df[vaccine_df.State!="India"]
vaccine
```

```python
vaccine = vaccine.rename(columns={"Total Individuals Vaccinated": "Total"})
vaccine
```

```python
#Picking top 5 Highest Vaccinated States

max_vac = vaccine.groupby(by="State")["Total"].sum().to_frame('Total').sort_values(by="Total", ascending = False)[:5]
# max_vac = max_vac[:5] works as well
```

```python
fig = plt.figure(figsize=(10,5))
maxi_vac = sns.barplot(data = max_vac.iloc[:10], y=max_vac["Total"],x=max_vac.index,linewidth=2,edgecolor="black")#y=max_vac.Total (works as well)
plt.title("Top 5 Most Vaccinated States", size=20)
plt.xlabel("States")
plt.ylabel("Total Vaccinated")
plt.show()
```

```python
#Top 5 Least Vaccinated States

min_vac = vaccine.groupby(by="State")["Total"].sum().to_frame("Total").sort_values(by = "Total")[:5]
min_vac
```

```python
fig = plt.figure(figsize=(15,9))
least_vac = sns.barplot(data = min_vac.iloc[:10],y=min_vac.Total,x=min_vac.index,linewidth=2,edgecolor="cyan")
plt.title("Top 5 Least Vaccinated States",size = 20)
plt.xlabel("State")
plt.ylabel("Total Vaccination")
plt.show()
```
